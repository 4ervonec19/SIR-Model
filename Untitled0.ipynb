```python
precip_type = object_stats['Precip Type']
precip_type_Serie = type(precip_type)
set_precip_type = list(set(list(precip_type))) # unique values of Precip Type
dictionary_precip_type = {}
for i in range(len(set_precip_type)):
  dictionary_precip_type[set_precip_type[i]] = i
dictionary_precip_type
#Work with types of precip - Ð¾ÑÐ°Ð´Ð¾Ðº
```

```python
summary = object_stats['Summary']
set_summary_type = list(set(list(summary))) # unique values of Summary
dictionary_summary_type = {}
for i in range(len(set_summary_type)):
  dictionary_summary_type[set_summary_type[i]] = i
dictionary_summary_type
```

```python

object_stats['Summary_numerical'] = object_stats['Summary'].map(dictionary_summary_type)
object_stats = object_stats.drop(columns = ['Summary'])
object_stats
```

```python
object_stats['Precip_Type_Numerical'] = object_stats['Precip Type'].map(dictionary_precip_type)
object_stats = object_stats.drop(columns = ['Precip Type'])
object_stats
```

```python
total_stats = total_stats.drop(columns = 'Apparent Temperature (C)')
total_stats
```

```python
target_value_temperature = total_stats['Temperature (C)']
target_value_temperature
```

```python
total_weather_parameters = total_stats.iloc[:, 1:]
total_weather_parameters
```

```python
total_weather_parameters.shape[0] == target_value_temperature.shape[0]
```

```python
total_stats["Temperature"] = total_stats['Temperature (C)']
total_stats
```

```python
data_correlation = total_stats.corr(method='pearson', numeric_only=True)

data_correlation
```

```python
data_correlation = data_correlation.fillna(0).round(2).reset_index(False)
data_correlation
```

```python
total_model_parameters = total_stats[['Humidity', 'Visibility (km)', 'Precip_Type_Numerical']]
total_model_parameters
```

```python
total_model_parameters.shape[0] == target_value_temperature.shape[0]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(total_model_parameters, target_value_temperature, test_size=0.3, random_state=42)
```

```python
model_weather_regression = LinearRegression()
model_weather_regression.fit(X_train, y_train)

predicted_values = model_weather_regression.predict(X_test)
```

```python
from sklearn import metrics
import numpy as np
print('Root mean Square Error:', np.sqrt(metrics.mean_squared_error(y_test, predicted_values)))
```

```python
print('R2: ', metrics.r2_score(y_test, predicted_values).round(2))
```

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.model_selection import cross_val_score


scores_data = pd.DataFrame()

alpha = np.linspace(0, 1, 30)
for ratio in alpha:

  model = Ridge(ratio)
  model.fit(X_train, y_train)
  train_score = metrics.r2_score(y_train, model.predict(X_train))
  test_score = metrics.r2_score(y_test, model.predict(X_test))
  cross_val = cross_val_score(model, X_train, y_train, cv = 5).mean()

  frame = pd.DataFrame({'alpha': [ratio] ,'train_score': [train_score], 'test_score': [test_score],
                        'cross_cal': [cross_val]})

  scores_data = scores_data.append(frame)






```
